```python
# processing requires a few libraries - these have to be installed in advance into the python environment
import xarray as xr
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import glob, os
```

```python
#data directory
datadir="./data/ERA5_rainfall_raw/"

files=glob.glob("{}/PRCPTOT_mon_ERA5_*_afghanistan_downloaded.nc".format(datadir))

outdir="./data/ERA5_rainfall/"
for file in files:
    print(file)
    ds=xr.open_dataset(file)
    for valid_time in ds["valid_time"]:
        da=ds.sel(valid_time=slice(valid_time,valid_time))
        year=valid_time.dt.year.data
        month=valid_time.dt.month.data

        outfile=f"{outdir}/tp_mon_ERA5_afghanistan_{year}{str(month).zfill(2)}.nc"
        if not os.path.exists(outfile):
            print(f"writing {outfile}")
            da.to_netcdf(outfile)
```
